```python
import pandas as pd
import numpy as np
```

```python
from pyecharts import Bar
bar = Bar('first bar','subTitle')
bar.add('clothes',["A","B","C","D","E"],[4,10,20,30,50])
bar
```

```python
from pyecharts import Line
line = Line('折线图')
line.add('clothes',["A","B","C","D","E"],[4,10,20,30,50])
line.add('pants',["A","B","C","D","E"],[4,10,20,30,30],is_fill=True,area_color='#000',area_opacity=0.3)
line
```

```python
from pyecharts import Pie
pie = Pie('饼图')
pie.add('clothes',["A","B","C","D","E"],[4,10,20,30,50],is_label_show = True)
pie
```

```python
from pyecharts import Scatter

v1 = [10, 20, 30, 40, 50, 60]
v2 = [10, 20, 30, 40, 50, 60]
scatter = Scatter("散点图示例")
scatter.add("A", v1, v2)
scatter.add("B", v1[::-1], v2 )

scatter
```

```python
index = pd.date_range('3/8/2017', periods=6, freq='M')
df1 = pd.DataFrame([1,2,3,4,5,6], index=index)
df2 = pd.DataFrame([1,2,3,4,5,7], index=index)

dtvalue1= [i[0] for i in df1.values]
dtvalue2 = df2[0].tolist()

bar = Bar('Bar chart', 'Profit and loss situation')
bar.add('profit', df1.index, dtvalue1,is_stack= True)
bar.add('loss', df2.index,  dtvalue2,is_stack= False,is_datazoom_show=True,is_convert=False, xaxis_interval=0, xaxis_rotate=30, yaxis_rotate=30)

bar
```

```python
from pyecharts import Funnel
df1 = pd.DataFrame({"rk":[100],"pf":[80],"ls":[60],"kc":[40]})
funnel = Funnel("零售漏斗图")
funnel.add("指标", df1.columns.tolist(), df1.values.tolist()[0], is_label_show=True, label_pos="inside", label_text_color="#fff")
funnel
```

```python
from pyecharts import Gauge
df1 = pd.DataFrame(np.array([1,2,3]),columns=['kpi'],index=['a','b','c'])
gauge = Gauge("零售指标图")
gauge.add("指标", "达成",df1.loc['b',])
gauge
```

```python
from pyecharts import Grid
grid = Grid()
grid.add(bar, grid_bottom="60%", grid_left="60%")
grid.add(scatter, grid_bottom="60%", grid_right="60%")
grid
```

```python
from pyecharts import Overlap
overlap = Overlap()
overlap.add(bar)
overlap.add(line)

overlap
```

```python
from pyecharts import WordCloud

name = ['a','b','c']
value = [10000, 6181, 4386]
wordcloud = WordCloud(width=600, height=600)
wordcloud.add("", name, value, word_size_range=[20, 100])

wordcloud
```

```python
from pyecharts import Map
from pyecharts import Map

value = [20, 190, 253, 77, 65]
attr = ['汕头市', '汕尾市', '揭阳市', '阳江市', '肇庆市']
map = Map("广东地图示例", width=1200, height=600)
map.add("", attr, value, maptype='广东', is_visualmap=True, visual_text_color='#000')

map
```
